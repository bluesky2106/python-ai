# Mining Stock data

### Task 1. Import the necessary libraries for your analysis

```python
import pandas as pd
```

### Task 2. Load dataset `stock.csv` into your notebook

```python
stock_df = pd.read_csv('stock.csv')
stock_df.head()
```

```python
stock_df.describe()
```

### Task 3. Copy this data to another object (another `DataFrame`) called `apple`

```python
apple = stock_df.copy(True)
```

### Task 4.  Check out the types of all columns

```python
apple.info()
```

```python
apple.dtypes
```

### Task 5. Let transform the values of `Date` column to `datetime` type

```python
apple.loc[:, 'Date'] = pd.to_datetime(apple['Date'])
```

```python
apple.info()
```

```python
apple['Date'].dtype
```

### Task 6. Re-index `apple` by `Date`

```python
apple = apple.set_index('Date')
apple
```

### Task 7.  Is there any duplicate dates in the dataset?

```python
apple.index
```

```python
apple.index.unique()
```

There are totally 8465 rows (count = 8465) and 8465 unique dates. Therefore, there is no duplicated dates in the dataset.

### Task 8.  If the index now is from the most recent date, please make the first entry the oldest date.

```python
apple = apple.sort_index(ascending=False)
apple
```

### Task 9. Extract the last business day of each month

```python
apple.loc[:, 'Date'] = apple.index
apple.loc[:, 'Year-Month'] = apple['Date'].apply(lambda x: x.strftime('%Y-%m'))
apple
```

```python
apple.groupby('Year-Month')['Date'].max()
```

### Task 10.  How many days are from the first day and the oldest?

```python
second = apple.index[0]
first = apple.index[len(apple.index)-1]
days = second - first
print('{} - {} = {} days'.format(second, first, days))
```

### Task 11.  How many months in the dataset you have been provided?

```python
apple.index.month.unique()
```

### Task 12. Plot the `Adj Close` value. Set the size of the figure to `13.5 x 9` inches

```python
%pylab inline
```

```python
figure(figsize=[13.5,9])
xlabel("Date")
ylabel("Adj Close")
title("Adj Close By Date")
apple['Adj Close'].resample('M').mean().plot()
```

### Task 13. Plot the averages of `Volume` and `Close` by month. 

```python
figure(figsize=[13.5,9])

subplot(121)
xlabel("Date")
ylabel("Volume")
title("Volume By Date")
apple['Volume'].resample('M').mean().plot()

subplot(122)
xlabel("Date")
ylabel("Close")
title("Close By Date")
apple['Close'].resample('M').mean().plot()
```

### Task 14. Select the subset of data from `01-01-2000` to `31-12-2013`.

```python
apple[(apple['Date'] >= '2000-01-01') & (apple['Date'] <= '2013-12-31')]
```

### Task 15. Add another column to the dataset called `Price Range`, which is the difference of `High` and `Low`.

```python
apple.loc[:, 'Price Range'] = apple['High'] - apple['Low']
apple
```

### Task 16. Add another column to the dataset called `Amount of Money`, which is the product of `Volume` and `Close`. 

```python
apple.loc[:, 'Amount of Money'] = apple['Volume'] * apple['Close']
apple
```
