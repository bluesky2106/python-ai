### Exercise 1
Define a function `max()`  that takes two numbers as arguments and returns the largest of them. Use the if­
then­else construct available in Python. (It is true that Python has the `max()`  function built in, but writing it
yourself is nevertheless a good exercise.). Remember to raise error if the inputs are not appropriate.

```
def max(a, b):
  if a > b:
    return a
  return b

print(max(2.0, 2.1))
print(max(3.0, 2.1))
```

### Exercise 2

Similarly, define a function `max_of_three()` that takes three numbers as arguments and returns the largest of
them.

```
def max_of_three(a, b, c):
  if a >= b and a >= c:
    return a
  elif b >= a and b >= c:
    return b
  return c

print(max_of_three(1.1, 1.1, 2))
print(max_of_three(1.1, 2.1, 2))
print(max_of_three(1.1, 2.1, 3.1))
print(max_of_three(1.1, 3.1, 2))
```

### Exercise 3

Define a function that computes the length of a given list or string. (It is true that Python has the `len()`
function built in, but writing it yourself is nevertheless a good exercise.). Remember to raise error if the inputs are not appropriate.

```
def len(a):
  count = 0
  try:
    for _ in a:
      count += 1
  except:
    raise TypeError('Wrong type')
  return count

print(len([2, 3]))
print(len('abc'))
print(len(2))
```

### Exercise 4

Write a function that takes a character (i.e. a string of length 1) and returns `True` if it is a vowel, `False`
otherwise.

```
def is_vowel(a):
  vowels = 'aeiou'
  if a in vowels:
    return True
  return False

print(is_vowel('e'))
print(is_vowel('c'))
```

### Exercise 5

Define a function `sum()` and a function `multiply()` that sums and multiplies (respectively) all the numbers in a list of numbers. For example, `sum([1, 2, 3, 4])` should return 10, and `multiply([1, 2, 3, 4])` should return 24.

```
def sum(lst):
  res = 0.0
  for v in lst:
    res += v
  return res

def multiply(lst):
  res = 1.0
  for v in lst:
    res *= v
  return res

print(sum([1,2,3,4]))
print(multiply([1,2,3,4]))
```

### Exercise 6

Define a function `reverse()` that computes the reversal
of a string. For example, `reverse("I am testing")`
should return the string `"gnitset ma I"`.

```
def reverse(s):
  res = ''
  for i in range(len(s)-1, -1, -1):
    res += s[i]
  return res

reverse("I am testing")
```

### Exercise 7

Define a function `is_palindrome()` that recognizes
palindromes (i.e. words that look the same written
backwards). For example, `is_palindrome("radar")`
should return `True`

```
def is_palindrome(s):
  ss = ''
  for i in range(len(s)-1, -1, -1):
    ss += s[i]
  return ss == s
```

```
print(is_palindrome('radar'))
print(is_palindrome('abc'))
```

### Exercise 8

Write a function `is_member()` that takes a value
(i.e. a number, string, etc) `x` and a list of
values `a`, and returns `True` if `x` is a member of `a`,
`False` otherwise.(Note that this is exactly what
the in operator does, but for the sake of the
exercise you should pretend Python did not
have this operator.)

```
def is_member(x, a):
  try:
    for v in a:
      if v == x:
        return True
  except:
    raise TypeError('Wrowng Type')
  return False
```

```
print(is_member(2, [1, 2]))
```

### Exercise 9

Define a function `overlapping()` that takes two lists
and returns `True` if they have at least one member
in common, `False` otherwise. You may use your
`is_member()` function, or the in operator, but for the
sake of the exercise, you should (also) write
it using two nested for-loops

```
def overlapping(lst1, lst2):
  for a in lst1:
    for b in lst2:
      if a == b:
        return True
  return False
```

```
print(overlapping([1,2], [2,3]))
print(overlapping([1,2], [4,3]))
```

### Exercise 10

Define a function `generate_n_chars()` that takes
an integer `n` and a character `c` and returns a string,
`n` characters long, consisting only of `c:s`. For
example, `generate_n_chars(5,"x")` should return
the string `"xxxxx"`. (Python is unusual
in that you can actually write an expression
`5 * "x"` that will evaluate to
`"xxxxx"`. For the sake of the exercise you should
ignore that the problem can
be solved in this manner.)

```
def generate_n_chars(n, c):
  res = ''
  for _ in range(n):
    res += c
  return res
```

```
print(generate_n_chars(5, 'x'))
```

### Exercise 12

The function `max()` from exercise 1) and the function
`max_of_three()` from exercise 2) will only work for
two and three numbers, respectively. But
suppose we have a much larger number of numbers,
or suppose we cannot tell in advance how many
they are? Write a function `max_in_list()` that takes a
list of numbers and returns the largest one

```
def max_in_list(lst):
  for i, v in enumerate(lst):
    if i == 0:
      res = v
    else:
      if res < v:
        res = v
  return res
```

```
max_in_list([1,2,3,4,7,5])
```

### Exercise 13

Write a function to extract numeric characters out of this type of string `[390996]培训店铺`, then convert it to integer.

```
import re

def extract_numeric(s):
  pattern = '\d+'
  find = re.findall(pattern, s)
  return int(find[0])

s = "[390996]培训店铺"
extract_numeric(s)
```

### Exercise 14

Write a function to extract id as 6 numeric characters (for example `392192`) out of these type of strings: `[392192]永定门外店`, `1030_392192_庄胜崇光`, `北京大兴国际机场店铺立地_392192`. Then please convert it to integer.

```
import re

def extract_id(s):
  pattern = '\d{6}'
  find = re.findall(pattern, s)
  return int(find[0])

ss = ["[392192]永定门外店", "1030_392192_庄胜崇光", "北京大兴国际机场店铺立地_392192"]
for s in ss:
  print(extract_id(s))
```

### Exercise 15

Write a function to convert this string `北京202105` to this string `2021-05`

```
import re

def convert(s):
  pattern = '\d{6}'
  find = re.findall(pattern, s)
  s = find[0]
  s = s[:4] + '-' + s[4:]
  return s

s = "北京202105"
print(convert(s))
```
