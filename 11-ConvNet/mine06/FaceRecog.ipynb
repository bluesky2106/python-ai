<a name='1'></a>
## 1 - Packages

```python
import cv2 
from face_recogizer import FaceRecognizer
```

<a name='2'></a>
## 2 - Register A New User

```python
fr = FaceRecognizer()
```

```python

# Automatic create 20 images for user Akagi
fr.register_face(user_name='VanNguyen', num_pictures=20)
```

<a name='3'></a>
## 3 - Encode New User

```python
encode_name = "encodings.pkl"
```

```python

fr.encode(encode_name)
```

<a name='4'></a>
## 4 - Recognize the new registered user

```python
encoding_dict = fr.load_pickle(encode_name)
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret,frame = cap.read()

    if not ret:
        print("CAM NOT OPEN") 
        break
    
    frame = fr.detect(frame, encoding_dict)

    cv2.imshow('camera', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```
