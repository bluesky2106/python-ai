# Face Recognition System

<a name='1'></a>
## 1 - Packages

```python
import os
import cv2
from face_recogizer import FaceRecognizer
```

<a name='2'></a>
## 2 - Register A New User

```python
fr = FaceRecognizer()

# Automatic create 20 images for user Akagi
fr.register_face(user_name='Akagi', num_pictures=30)
```

<a name='3'></a>
## 3 - Encode New User

```python
encode_name = "encodings.pkl"
fr = FaceRecognizer()
fr.encode(encode_name)
```

<a name='4'></a>
## 4 - Recognize user

```python
encoding_dict = fr.load_pickle(encode_name)
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret,frame = cap.read()

    if not ret:
        print("CAM NOT OPEN") 
        break
    
    frame = fr.detect(frame, encoding_dict)

    cv2.imshow('camera', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

<a name='5'></a>
## 5 - Evaluate

```python
encode_name = "encodings.pkl"
fr = FaceRecognizer()
encoding_dict = fr.load_pickle(encode_name)
```

```python
correct = 0
wrong = 0
for label in os.listdir('faces'):
    for image_name in os.listdir(os.path.join('faces', label)):
        # print(image_name)
        img_BGR = cv2.imread(os.path.join('faces', label, image_name))
        img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)
        name = fr.detect_name(img_RGB, encoding_dict)
        if name == label:
            correct += 1
        else:
            wrong += 1

accuracy = correct / (correct + wrong)
print('Accuracy:', accuracy)
```
