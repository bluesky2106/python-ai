```python
import numpy 
import pandas as pd
from sklearn.model_selection import train_test_split #thư viện tách dữ liệu
from sklearn.svm import SVC 
import joblib
import time
```

```python
data = pd.read_csv("train.tsv",quoting = 3, delimiter = "\t", header= 0)

data.drop(["PhraseId", "SentenceId"],axis=1, inplace=True) # xóa cột PhraseId và SentenceId

data = data[:5000] # chọn ra 1000 dữ liệu đầu tiên
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(data["Phrase"]).toarray()
X
```

```python
y = data.Sentiment.values
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0) # chia thành 2 tập training (80%) và testing (20%)
```

# Default kernel (rbf)

```python
start_time = time.time()
clf = SVC()

clf_rbf = clf.fit(X_train, y_train)

print('Accuraycy: ',clf_rbf.score(X_test, y_test))
print('Time to train model: ', time.time() - start_time, 's.')
```

# Linear kernel

```python
start_time = time.time()

clf = SVC(kernel='linear')

clf_linear = clf.fit(X_train, y_train)

print('Accuraycy: ',clf_linear.score(X_test, y_test))
print('Time to train model: ', time.time() - start_time, 's.')
```

# Poly kernel

```python
start_time = time.time()

clf = SVC(kernel='poly')

clf_poly = clf.fit(X_train, y_train)

print('Accuraycy: ',clf_poly.score(X_test, y_test))
print('Time to train model: ', time.time() - start_time, 's.')
```

# Conclusion

Time training: Linear kernel is better than Non-linear

Accuracy: Non-linear kernel is better than Linear
