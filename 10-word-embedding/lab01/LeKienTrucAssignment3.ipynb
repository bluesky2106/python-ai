```python
import os
import pandas as pd
import string
import gensim
from gensim.models import Word2Vec
from gensim.test.utils import common_texts
from gensim.test.utils import get_tmpfile
```

```python
data = pd.read_csv('vne_content_title_dulich.csv')
data
```

```python

sentences = []
for sentence in data["content"]:
  ss = sentence.split(".")
  if not ss:
    continue
  for s1 in ss:
    sentences.append(s1.split())
```

```python
sentences
```

```python
model = Word2Vec(sentences, min_count = 1, window = 5, sg = 1, workers = 5)
model.wv.save('word2vec_skipgram.model')
model.save('w2v.model')
```

```python
new_model = gensim.models.KeyedVectors.load('word2vec_skipgram.model')
```

```python
def test_result(word, new_model):
  word = word.replace(" ", "_")
  word = word.lower()
  return new_model.most_similar(positive=[word],topn=10)
```

```python
test_result("khách sạn", new_model)
```

```python
test_result("giải trí", new_model)
```
